```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_csv("hf://datasets/xuejun72/HR-VILAGE-3K3M/study_meta.csv")
df.head()
```

```python
df.to_csv('study_meta.csv', index=False)
print("DataFrame saved as study_meta.csv")
```

```python
df.sample(10)
```

```python
original_rows = df.shape[0]
df_filtered = df[(df["num_subjects"] > 20) & (df["num_observations"] > 100) & (df["study_type"] == "vaccine")]
filtered_rows = df_filtered.shape[0]
print(f"  : {original_rows}")
print(f"   : {filtered_rows}")
display(df_filtered)
```

```python
print(len(df_filtered))
latest_df = df_filtered.drop_duplicates()
print(len(latest_df))
```

```python
latest_df["genes_per_subject"] = latest_df["num_observed_genes"] / latest_df["num_subjects"]
latest_df[["vaccine", "num_observed_genes", "num_subjects", "genes_per_subject"]].head()
latest_df
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(
    data=latest_df,
    x="vaccine",
    y="genes_per_subject",
    showmeans=True,
    meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black"}
)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Observed Genes per Subject")
plt.xlabel("Vaccine")
plt.title("Distribution of Observed Genes per Subject by Vaccine")
plt.tight_layout()
plt.savefig("genes_per_subject_boxplot.png", dpi=300)
plt.show()
```
